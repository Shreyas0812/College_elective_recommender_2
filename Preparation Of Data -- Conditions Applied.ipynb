### Getting Data from the user: 

```python
from google.colab import files

uploaded = files.upload()

file_name = list(uploaded.keys())[0]
```

```python
import random
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(file_name, sheet_name='Sheet1')
df.head()
```

```python
df.head(10)
```

```python
Semester = 'Sem_6'
marks_thresh = 65
```

```python
df_sem = df[df['Semster_Name'] == Semester].reset_index(drop=True)
df_sem.head(10)
```

```python
#Applying Marks Threshold
df_sem_marks_thresh = df_sem[df_sem['Rating'] > marks_thresh].reset_index(drop=True)
df_sem_marks_thresh.head()
```

```python
df_sem_marks_thresh
```

```python
def random_sum_to(n, num_terms):
    new_num = int(n/num_terms)  
    a = random.sample(range(new_num, 21), num_terms)
    
    remaining_num = int((sum(a) - n)/5)
    carry_over = (sum(a) - n)%5

    for i in range(len(a)):
      a[i] = a[i] - remaining_num

    random_idx = random.randint(0,4)
    a[random_idx] = a[random_idx] - carry_over

    return a
```

```python
for index, row in df_sem_marks_thresh.iterrows():

  row_data = random_sum_to(df_sem_marks_thresh.loc[index, 'Rating'], 5)

  df_sem_marks_thresh.loc[index, 'subject_content'] = row_data[0]
  df_sem_marks_thresh.loc[index, 'teacher_rating'] = row_data[1]
  df_sem_marks_thresh.loc[index, 'subject_interest'] = row_data[2]
  df_sem_marks_thresh.loc[index, 'practical_knowledge'] = row_data[3]
  df_sem_marks_thresh.loc[index, 'likely_to_recommend'] = row_data[4]
```
